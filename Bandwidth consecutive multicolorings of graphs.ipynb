## I. Lý thuyết

1. **Đề bài.**
    - Đồ thị $G(V, E)$ là đồ thị có $V$ đỉnh và $E$ cạnh.
    - $F(v)$ là `dãy số nguyên dương` liên tiếp được gán cho đỉnh $v$.
    - $f(v)$ là số nguyên lớn nhất của $F(v)$
    - $b(v_1, v_2) + 1$ là trọng số `nguyên dương` của cạnh $(v_1, v_2)$ đại diện cho khoảng cách ngắn nhất của 2 dãy số $F(v_1)$ và $F(v_2)$.
    - $b(v)$ là `trọng số nguyên dương` của đỉnh đại diện cho số lượng số nguyên dương trong $F(v)$.
    - $span = N$ là một phép tô màu F, $span$ là số nguyên lớn nhất được gán cho bất kì đỉnh nào của đồ thị
    - $χ_b(G)$: Là giá trị span nhỏ nhất có thể đạt được trong tất cả các phép tô màu hợp lệ

> Cho $G$ là một đồ thị đơn, trong đó mỗi đỉnh $v$ được gán một trọng số nguyên dương $b(v)$, và mỗi cạnh $(v_1,v_2)$ được gán một trọng số nguyên không âm $b(v_1,v_2)$. Một `bandwidth consecutive multicoloring` của $G$ là một cách gán cho mỗi đỉnh $v$ đúng $b(v)$ số nguyên dương liên tiếp, sao cho **với mỗi cạnh $(v_1,v_2)$, mọi số nguyên được gán cho đỉnh $v$ đều khác mọi số nguyên được gán cho đỉnh $w$ một lượng lớn hơn $b(v,w)$**. Số nguyên lớn nhất được gán cho bất kỳ đỉnh nào được gọi là span (độ trải) của phép tô màu.
> 
> $span(F) = max_{v \in V}f(v)$ \
> $χ_b(G) = min_F\{span(F)\}$

## II. Ràng buộc

1. **Ràng buộc màu dương**: Để đảm bảo các màu là `số nguyên dương`
> $b(v) \le f(v)$
3. **Ràng buộc khác biệt và duy nhất**: Vì $b(v_1, v_2) \ge 0$, giá trị màu lớn nhất của hai đỉnh kề nhau không được trùng nhau và mỗi đỉnh `chỉ có một` $F(v)$ hay $f(v)$ duy nhất.
> $f(v_2) \ne f(v_1)$
4. **Ràng buộc khoảng cách**: Khoảng cách giữa các băng thông phải lớn hơn $b(v_1, v_2)$
> $f(v_2) - f(v_1) > b(v_2) - 1 + b(v_1, v_2)$ ($b(v_2) - 1$ là vì số phần tử là $b(v_2)$ thì số khoảng là $b(v_2) - 1$) \
> $\implies f(v_2) - f(v_1) \ge b(v_2) + b(v_1, v_2)$

## III. Mã hóa

### A. Order Encoding with Directional Variables

1. **Biến:**
    - $X_{v, i} = 1$ $\iff$ $f(v) \ge i$
    - $D_{v_1, v_2}= 1 \iff f(v_1) > f(v_2)$
    > Nếu đỉnh $v$ có `số nguyên dương` lớn nhất $f(v) \ge i$ thì $X_{v,i} = 1$

2. **Encode** 
    - Ràng buộc tính liên tục:
        > $$\bigwedge_{v \in V} \ \bigwedge_{i = b(i)} ^ {N - 1}X_{v, i + 1} \rightarrow X_{v, i}$$
        > $$\equiv \bigwedge_{v \in V} \ \bigwedge_{i = b(i)} ^ {N - 1} \neg X_{v, i + 1} \lor X_{v, i}$$
        > - Nếu đỉnh $v$ có $f(v)$ lớn hơn hoặc bằng $i + 1$ thì chắc chắn $f(v)$ cũng lớn hơn hoặc bằng $i$
    - Ràng buộc biên trên (luôn luôn lớn hơn hoặc bằng $b(v)$)
        > $$\bigwedge_{v \in V} X_{v, b(v)}$$
    - Ràng buộc biên dưới (luôn luôn không vượt quá $span = N$)
        > $$\bigwedge_{v \in V}\neg X_{v, N + 1}$$
    
    - Ràng buộc khoảng cách: 
        - TH1: $f(v_1) > f(v_2)$
            > $$f(v_1) \ge f(v_2) + C$$ trong đó $$C = b(v_1) + b(v_1, v_2)$$
            > $$\implies f(v_2) \ge i \rightarrow f(v_1) \ge i + C$$
            > $$\equiv \bigwedge_{(v_1, v_2) \in E}\ \bigwedge_{i = max(b(v_1), b(v_2) - C)} ^ {N - C}  D_{v_1, v_2} \rightarrow( X_{v_2, i} \rightarrow X_{v_1, i + C})$$ 
            > $$\equiv \bigwedge_{(v_1, v_2) \in E}\ \bigwedge_{i = max(b(v_1), b(v_2) - C)} ^ {N - C}   \neg D_{v_1, v_2}\lor \neg X_{v_2, i} \lor X_{v_1, i + C}$$
    
        - TH2: $f(v_2) > f(v_1)$
            > $$f(v_2) \ge f(v_1) + C$$ trong đó $$C = b(v_2) + b(v_1, v_2)$$
            > $$\implies f(v_1) \ge i \rightarrow f(v_2) \ge i + C$$
            > $$\equiv \bigwedge_{(v_1, v_2) \in E}\ \bigwedge_{i = max(b(v_1), b(v_2) - C)} ^ {N - C} \neg D_{v_1, v_2} \rightarrow( X_{v_1, i} \rightarrow X_{v_2, i + C})$$ 
            > $$\equiv \bigwedge_{(v_1, v_2) \in E}\ \bigwedge_{i = max(b(v_1), b(v_2) - C)} ^ {N - C}   D_{v_2, v_1}\lor \neg X_{v_1, i} \lor X_{v_2, i + C}$$


3. **Độ phức tạp**
    - Số lượng biến:
        - $X_{v, i}$ trong đó $v \in V$ và $i \in \{b(v), \dots, N \}$ $\implies$ số lượng biến X là $O(|V|\cdot N)$
        - $D_{v_1, v_2}$ trong đó $(v_1, v_2) \in E \implies$ số lượng biến là $O(|E|)$
          > **Tổng số biến là** $O(|V| \cdot N + |E|)$ 
    - Số lượng mệnh đề:
        - $\displaystyle \bigwedge_{v \in V} \ \bigwedge_{i = b(i)} ^ {N - 1} \neg X_{v, i + 1} \lor X_{v, i} \implies$ số lượng mệnh đề là $O(|V|\cdot N)$
        - $\displaystyle\bigwedge_{v \in V} X_{v, b(v)} \implies$ số lượng mệnh đề là $O(|V|)$
        - $\displaystyle\bigwedge_{v \in V}\neg X_{v, N + 1} \implies$ số lượng mệnh đề là $O(|V|)$
        - $\displaystyle\bigwedge_{(v_1, v_2) \in E}\ \bigwedge_{i = max(b(v_1), b(v_2) - C)} ^ {N - C}   \neg D_{v_1, v_2}\lor \neg X_{v_2, i} \lor X_{v_1, i + C} \implies$ số lượng mệnh đề $O(|E|\cdot N)$
        - $\displaystyle\bigwedge_{(v_1, v_2) \in E}\ \bigwedge_{i = max(b(v_1), b(v_2) - C)} ^ {N - C}   D_{v_2, v_1}\lor \neg X_{v_1, i} \lor X_{v_2, i + C} \implies$ số lượng mệnh đề $O(|E|\cdot N)$
          > **Tổng số mệnh đề là** $O((|E|+|V|)\cdot N)$

### B.Order Encoding without Directional Variables

1. **Biến**
    - $X_{v, i} = 1 \iff f(v)\ge i$
    - $Y_{v, i} = 1 \iff f(v) = i$

2. **Encode**
    - Ràng buộc tính liên tục:
        > $$\bigwedge_{i = 1} ^ {N - 1}X_{v, i + 1} \rightarrow X_{v, i}$$
        > $$\equiv \bigwedge_{i = 1} ^ {N - 1} \neg X_{v, i + 1} \lor X_{v, i}$$
        > - Nếu đỉnh $v$ có $f(v)$ lớn hơn hoặc bằng $i + 1$ thì chắc chắn $f(v)$ cũng lớn hơn hoặc bằng $i$
    - Ràng buộc biên trên (luôn luôn lớn hơn hoặc bằng 1)
        > $$\bigwedge_{v \in V} X_{v, b(v)}$$
    - Ràng buộc biên dưới (luôn luôn không vượt quá $span = N$)
        > $$\bigwedge_{v \in V} \neg X_{v, N + 1}$$
    - Ràng buộc giữa $X$ và $Y$
      > $$Y_{v,i} \leftrightarrow X_{v, i} \land \neg X_{v, i + 1}$$
    - Ràng buộc khoảng cách
      > $$Y_{v, i} \rightarrow (X_{u, i  + C} \lor \neg X_{u, i - C + 1})$$

### C. Ladder Encoding

1. **Biến**
    - $X_{v,i} = 1 \iff$ đỉnh $v$ được gán nhãn $f(v) = i$
   - $D_{v_1, v_2}= 1 \iff f(v_1) > f(v_2)$


2. **Encode**
    - Mỗi đỉnh chỉ được gán 1 nhãn $i$
      > $$\bigwedge_{v \in V} \ \sum_{i = b(i)}^{N}X_{v,i} = 1$$
    - Ràng buộc khoảng cách
      > $$
        d_{uv}=b(u,v)+b(v),
        \qquad
        d_{vu}=b(u,v)+b(u).
        $$
        - TH1: $f(v) > f(u)$
          > $$
            \bigwedge_{(u,v) \in E}(D_{uv}
            \rightarrow
            \bigwedge_{s=max(b(u), b(v))}^{N-d_{uv}}
            \left(
            \sum_{l=s}^{s+d_{uv}}
            (X_{u,l}+X_{v,l})
            \le1
            \right))
            $$
      - TH2: $f(u) > f(v)$
        > $$
         \bigwedge_{(u,v) \in E}(
            \neg D_{uv}
            \rightarrow
            \bigwedge_{s=max(b(u), b(v))}^{N-d_{vu}}
            \left(
            \sum_{l=s}^{s+d_{vu}}
            (X_{u,l}+X_{v,l})
            \le1
            \right))
            $$

3. **Độ phức tạp**
    - Số lượng biến:
        - $X_{v,i} \implies$ số lượng biến là $O(|V|\cdot N)$
        - $D_{v_1, v_2} \implies$ số lượng biến là $O(|E|)$
          > **Tổng số biến là** $O(|V|\cdot N + |E|)$
    - Số lượng mệnh đề:
        - $\displaystyle\bigwedge_{v \in V} \ \sum_{i = b(i)}^{span}X_{v,i} = 1$
          $\displaystyle\implies$ số lượng mệnh đề là $O(N)$
        - $\displaystyle\bigwedge_{(u,v) \in E}\ (D_{uv}\rightarrow\bigwedge_{s=max(b(u), b(v))}^{N-d_{uv}}\left(\sum_{l=s}^{s+d_{uv}}(X_{u,l}+X_{v,l})\le1\right)) \implies$ số lượng mệnh đề là $O(|V|\cdot N)$
        - $\displaystyle\bigwedge_{(u,v) \in E} \ (\neg D_{uv}\rightarrow\bigwedge_{s=max(b(u), b(v))}^{N-d_{vu}}\left(\sum_{l=s}^{s+d_{vu}}(X_{u,l}+X_{v,l})\le1\right)) \implies$ số lượng mệnh đề là $O(|V|\cdot N)$
          > **Tổng số mệnh đề là** $O(|V|\cdot N)$

3. **Độ phức tạp**

    * Số lượng biến:
    
      * $X_{v,i}$ $\implies$ số lượng biến là $O(|V| \cdot N)$.
      * $Y_{v,i}$ trong đó $v \in V$ và $i \in {b(v), \dots, N}$ $\implies$ số lượng biến là $O(|V| \cdot N)$.
    
        > **Tổng số lượng là** $O(2|V| \cdot N)$.
    
    * Số lượng mệnh đề:
      * $\displaystyle \bigwedge_{v \in V}\ \bigwedge_{i=b(v)}^{N-1}\left(\neg X_{v,i+1}\lor X_{v,i}\right)$
        $\implies$ số lượng mệnh đề là $O(|V| \cdot N)$.
      * $\displaystyle \bigwedge_{v \in V} X_{v,b(v)}$
        $\implies$ số lượng mệnh đề là $O(|V|)$.
      * $\displaystyle \bigwedge_{v \in V}\neg X_{v,N+1}$
        $\implies$ số lượng mệnh đề là $O(|V|)$.
      * $\displaystyle \bigwedge_{v \in V}\ \bigwedge_{i=b(v)}^{N}\left(Y_{v,i}\leftrightarrow X_{v,i}\land\neg X_{v,i+1}\right)$
        $\implies$ mỗi phép tương đương được chuyển sang $3$ mệnh đề CNF, nên số lượng mệnh đề là $O(|V| \cdot N)$.
      * $\displaystyle \bigwedge_{(u,v)\in E}\ \bigwedge_{i}\left(Y_{v,i}\rightarrow(X_{u,i+C}\lor\neg X_{u,i-C+1})\right)$
        $\implies$ với mỗi cạnh có $O(N)$ giá trị của $i$, nên số lượng mệnh đề là $O(|E| \cdot N)$.
        > **Tổng số mệnh đề là** $O((|V| + |E|)\cdot N)$.

## IV.Đánh giá độ phức tạp

### A. Order Encoding with Directional Variables

| Tiêu chí | Độ phức tạp |
|----------|-------------|
| **Số lượng biến** | $O(|V| \cdot N + |E|)$ |
| **Số lượng mệnh đề** | $O((|V| + |E|)\cdot N)$ |

**Đánh giá**

- Là phương pháp cân bằng nhưng phụ thuộc tuyến tính vào cả số đỉnh và số cạnh ở cả hai tiêu chí.
- **Đồ thị thưa** ($|E| \approx |V|$):
  - Hoạt động rất tốt.
  - Cả số biến và số mệnh đề đều đạt mức:
    $$
    O(|V|\cdot N)
    $$
- **Đồ thị dày/đầy đủ** ($|E| \approx |V|^2$):
  - Số biến và số mệnh đề đều tăng mạnh do phụ thuộc vào $|E|$.
  - Độ phức tạp tiến tới:
    - Biến: $O(|V|^2)$ (ngoài thành phần $|V|N$)
    - Mệnh đề: $O(|V|^2N)$





### B. Order Encoding without Directional Variables

| Tiêu chí | Độ phức tạp |
|----------|-------------|
| **Số lượng biến** | $O(2|V|\cdot N)\sim O(|V|\cdot N)$ |
| **Số lượng mệnh đề** | $O((|V|+|E|)\cdot N)$ |

**Đánh giá**

- Ưu điểm
    - Số lượng biến hoàn toàn **không phụ thuộc** vào số cạnh $|E|$.
    - Khi đồ thị rất dày ($|E| \gg |V|$), đây là lợi thế đáng kể về bộ nhớ.

- Hạn chế
    - Số lượng mệnh đề vẫn phụ thuộc vào $|E|$.
    - Với đồ thị đầy đủ:
  $$
  |E|\approx |V|^2,
  $$
  số mệnh đề vẫn tăng rất lớn.





### C. Ladder Encoding

| Tiêu chí | Độ phức tạp |
|----------|-------------|
| **Số lượng biến** | $O(|V|\cdot N+|E|)$ |
| **Số lượng mệnh đề** | $O(2|V|\cdot N)\sim O(|V|\cdot N)$ |

#### Đánh giá

**Ưu điểm**

- Số lượng mệnh đề hoàn toàn **không phụ thuộc** vào số cạnh $|E|$.
- Với đồ thị dày hoặc đầy đủ:
  $$
  |E|\approx |V|^2,
  $$
  số lượng mệnh đề vẫn duy trì ở mức:
  $$
  O(|V|\cdot N).
  $$
- Đây là lợi thế lớn đối với SAT solver vì số lượng mệnh đề thường ảnh hưởng trực tiếp đến hiệu năng giải.

**Hạn chế**

- Số lượng biến tăng tuyến tính theo số cạnh:
  $$
  O(|V|\cdot N+|E|).
  $$

### C. Ladder Encoding

| Tiêu chí | Độ phức tạp |
|----------|-------------|
| **Số lượng biến** | $O(|V|\cdot N+|E|)$ |
| **Số lượng mệnh đề** | $O(2|V|\cdot N)\sim O(|V|\cdot N)$ |

**Đánh giá**

- Ưu điểm
    - Số lượng mệnh đề hoàn toàn **không phụ thuộc** vào số cạnh $|E|$.
    - Với đồ thị dày hoặc đầy đủ:
      $$
      |E|\approx |V|^2,
      $$
      số lượng mệnh đề vẫn duy trì ở mức:
      $$
      O(|V|\cdot N).
      $$
    - Đây là lợi thế lớn đối với SAT solver vì số lượng mệnh đề thường ảnh hưởng trực tiếp đến hiệu năng giải.

- Hạn chế
    - Số lượng biến tăng tuyến tính theo số cạnh:
      $$
      O(|V|\cdot N+|E|).
      $$


### Tổng kết

| Encoding | Số biến | Số mệnh đề | Phù hợp nhất |
|----------|----------|------------|--------------|
| **Order Encoding (with directional variables)** | $O(|V|N+|E|)$ | $O((|V|+|E|)N)$ | Đồ thị thưa, cân bằng giữa biến và mệnh đề |
| **Order Encoding (without directional variables)** | $O(|V|N)$ | $O((|V|+|E|)N)$ | Khi muốn giảm số lượng biến |
| **Ladder Encoding** | $O(|V|N+|E|)$ | $O(|V|N)$ | Đồ thị dày hoặc đầy đủ, ưu tiên giảm số lượng mệnh đề |
