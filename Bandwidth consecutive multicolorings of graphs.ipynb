# Bandwidth consecutive multicolorings of graphs

## I. Lý thuyết

**1. Đề bài.**
- Đồ thị $G(V, E)$ là đồ thị có $V$ đỉnh và $E$ cạnh.
- $F(v)$ là `dãy số nguyên dương` liên tiếp được gán cho đỉnh $v$.
- $f(v)$ là số nguyên lớn nhất của $F(v)$
- $b(v_1, v_2) + 1$ là trọng số `nguyên dương` của cạnh $(v_1, v_2)$ đại diện cho khoảng cách ngắn nhất của 2 dãy số $F(v_1)$ và $F(v_2)$.
- $b(v)$ là `trọng số nguyên dương` của đỉnh đại diện cho số lượng số nguyên dương trong $F(v)$.
- $span$ là một phép tô màu F, $span$ là số nguyên lớn nhất được gán cho bất kì đỉnh nào của đồ thị
- $χ_b(G)$: Là giá trị span nhỏ nhất có thể đạt được trong tất cả các phép tô màu hợp lệ

> Mục tiêu của bài toán $BCM$ là tối hiểu hóa giá trị $span$ trên đồ thị
> $span(F) = max_{v \in V}f(v)$ \
> $χ_b(G) = min_F\{span(F)\}$

## II. Ràng buộc

1. **Ràng buộc màu dương**: Để đảm bảo các màu là `số nguyên dương`
> $b(v) \le f(v)$
3. **Ràng buộc khác biệt và duy nhất**: Vì $b(v_1, v_2) \ge 0$, giá trị màu lớn nhất của hai đỉnh kề nhau không được trùng nhau và mỗi đỉnh `chỉ có một` $F(v)$ hay $f(v)$ duy nhất.
> $f(v) \ne f(v_1)$
4. **Ràng buộc khoảng cách**: Khoảng cách giữa các băng thông phải lớn hơn $b(v_1, v_2)$
> $f(v_2) - f(v_1) > b(v_2) - 1 + b(v_1, v_2)$ ($b(v_2) - 1$ là vì số phần tử là $b(v_2)$ thì số khoảng là $b(v_2) - 1$) \
> $\implies f(v_2) - f(v_1) \ge b(v_2) + b(v_1, v_2)$ 

## III. Mã hóa

### A. Theo hướng order encoding

**1. Biến:**
- $X_{v, i} = 1$ $\iff$ $f(v) \ge i$
- $D_{v_1, v_2}= 1 \iff f(v_1) > f(v_2)$
> Nếu đỉnh $v$ có `số nguyên dương` lớn nhất $f(v) \ge i$ thì $X_{v,i} = 1$

**2. Encode** 
- Ràng buộc tính liên tục:
    > $$\bigwedge_{i = 1} ^ {N - 1}X_{v, i + 1} \rightarrow X_{v, i}$$
    > $$\equiv \bigwedge_{i = 1} ^ {N - 1} \neg X_{v, i + 1} \lor X_{v, i}$$
    > - Nếu đỉnh $v$ có $f(v)$ lớn hơn hoặc bằng $i + 1$ thì chắc chắn $f(v)$ cũng lớn hơn hoặc bằng $i$
- Ràng buộc biên trên (luôn luôn lớn hơn hoặc bằng 1)
    > $$X_{v, 1}$$
- Ràng buộc biên dưới (luôn luôn không vượt quá $span = N$)
    > $$\neg X_{v, N + 1}$$

- Ràng buộc khoảng cách: 
    - TH1: $f(v_1) > f(v_2)$
        > $$f(v_1) \ge f(v_2) + C$$ trong đó $$C = b(v_1) + b(v_1, v_2)$$
        > $$\implies f(v_2) \ge i \rightarrow f(v_1) \ge i + C$$
        > $$\equiv \bigwedge_{(v_1, v_2) \in E}\bigwedge_{i = 1} ^ {N - C}  D_{v_1, v_2} \rightarrow( X_{v_2, i} \rightarrow X_{v_1, i + C})$$ 
        > $$\equiv \bigwedge_{v \in V}\bigwedge_{i = 1} ^ {N - C}   \neg D_{v_1, v_2}\lor \neg X_{v_2, i} \lor X_{v_1, i + C}$$

    - TH2: $f(v_2) > f(v_1)$
        > $$f(v_2) \ge f(v_1) + C$$ trong đó $$C = b(v_2) + b(v_1, v_2)$$
        > $$\implies f(v_1) \ge i \rightarrow f(v_2) \ge i + C$$
        > $$\equiv \bigwedge_{(v_1, v_2) \in E}\bigwedge_{i = 1} ^ {N - C} \neg D_{v_1, v_2} \rightarrow( X_{v_1, i} \rightarrow X_{v_2, i + C})$$ 
        > $$\equiv \bigwedge_{v \in V}\bigwedge_{i = 1} ^ {N - C}   D_{v_2, v_1}\lor \neg X_{v_1, i} \lor X_{v_2, i + C}$$


### B. Theo hướng sequential encoding 

1. **Biến**
    - $X_{v,i} = 1 \iff$ đỉnh $v$ được gán nhãn $f(v) = i$
   - $D_{v_1, v_2}= 1 \iff f(v_1) > f(v_2)$

2. **Encode**
    - Mỗi đỉnh chỉ được gán 1 nhãn $i$
      > $$\bigwedge_{v \in V} \ \sum_{i = b(i)}^{span}X_{v,i} = 1$$
    - Ràng buộc khoảng cách
      > $$
        d_{uv}=b(u,v)+b(v),
        \qquad
        d_{vu}=b(u,v)+b(u).
        $$
        - TH1: $f(v_1) > f(v_2)$
          > $$
            D_{uv}
            \rightarrow
            \bigwedge_{s=1}^{K-d_{uv}}
            \left(
            \sum_{l=s}^{s+d_{uv}}
            (X_{u,l}+X_{v,l})
            \le1
            \right).
            $$
      - TH2: $f(v_2) > f(v_1)$
        > $$
            \neg D_{uv}
            \rightarrow
            \bigwedge_{s=1}^{K-d_{vu}}
            \left(
            \sum_{l=s}^{s+d_{vu}}
            (X_{u,l}+X_{v,l})
            \le1
            \right).
            $$
